```python
import pandas as pd
from pandas import DataFrame
import zipfile

train = DataFrame()

with zipfile.ZipFile('train.zip') as existing_zip:
    existing_zip.extractall('./')
    for train_file in existing_zip.namelist():
        tmp = pd.read_csv(train_file, sep='\t')
        train = pd.concat([train, tmp], axis=0)

print(train.shape)
```

```python
train
```

```python
train['user_ids'] = train['user_id'].apply(lambda x: x.split('_')[0])
train['user_type'] = train['user_id'].apply(lambda x: x.split('_')[1])
train['product_ids'] = train['product_id'].apply(lambda x: x.split('_')[0])
train['product_type'] = train['product_id'].apply(lambda x: x.split('_')[1])
```
